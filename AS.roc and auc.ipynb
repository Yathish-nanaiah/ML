```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def model1():

    con1=np.ndarray([2,2])
    xa,ya=(4,4)  #xa=x actual, ya=y actual
    xp,yp=(2,6)  #xp=x predicted, yp=y predicted
    t=xa+ya

    fpr1=np.array([10])  #false positive rate
    tpr1=np.array([10])  #true positive rate

    while(1):
        if(xp<=xa):
            tp=xp
            tp=tp*tp
            tp=np.sqrt(tp)
            con1[0][0]=tp    
        else:
            tp=xa
            tp=tp*tp
            tp=np.sqrt(tp)
            con1[0][0]=tp
        if(yp<=ya):
            tn=yp
            tn=tn*tn
            tn=np.sqrt(tn)
            con1[1][1]=tn
        else:
            tn=ya
            tn=tn*tn
            tn=np.sqrt(tn)
            con1[1][1]=tn
        if(yp<=xp):
            fn=0
            con1[0][1]=fn
        else:
            fn=xp-xa
            fn=fn*fn
            fn=np.sqrt(fn)
            con1[0][1]=fn
        if(yp<=xp):
            fp=yp-ya
            tp=tp*tp
            tp=np.sqrt(tp)
            con1[1][0]=fp
        else:
            fp=con1[1][0]=0
        break
    print("confusion matrix1")
    print(con1)                                
    print(" ")
    if(fp==0):
        fpr1=0
    else:
        fpr1=fp/(fp+tn)
    print("fpr1= ", fpr1)
    print(" ")
    if(tp==0):
        tpr1=0
    else:    
        tpr1=tp/(tp+fn)
    print("tpr1= ", tpr1)

    return(tpr1,fpr1)
```
